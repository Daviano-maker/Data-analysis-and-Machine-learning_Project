## Loading Multiple Log LAS

```python
import lasio
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
las_file_list = []
path = "Documents/Yinka's PHD data/"
```

```python
files = os.listdir(path)
files
```

```python
#reading the las files; how to select only the LAS files
for file in files:
    if file.lower().endswith('.las'):
        las_file_list.append(path + file)
        
las_file_list
```

### Appending the LAS files to a Dataframe

```python
df_list = []

for lasfile in las_file_list:
    las = lasio.read(lasfile)
    lasdf = las.df()
    
    lasdf['WELL'] = las.well.WELL.value
    lasdf['DEPTH'] = lasdf.index
    df_list.append(lasdf)
```

```python
EBDdf = pd.concat(df_list, sort=True)
```

## Plot all well logs for all wells

```python
import os
import sys
import numpy as np
import copy
from tkinter import *
from tkinter import filedialog
import ipywidgets as widgets
from ipywidgets import interact
```

```python
def logs_plot_per_well(logs):
    '''
    Function to plot well logs
    '''

    well_name = EBDdf_2["WELL"].unique().tolist()
    log = copy.deepcopy(logs)
    try:
        log.drop(['DEPTH','WELL'], axis='columns', inplace=True)
    except:
        log.drop(['DEPT','WELL'], axis='columns', inplace=True)
    
    fig, ax=plt.subplots(nrows=len(well_name), ncols=len(log.columns), figsize=(30,40), squeeze=False)
    for i in range(len(log.columns)):
        for j in range(len(well_name)):
            figure = ax[j][i]
            try:
                Depth = logs[logs['WELL']==well_name[j]].DEPTH
            except:
                Depth = logs[logs['WELL']==well_name[j]].DEPT
            figure.plot(log[logs['WELL']==well_name[j]].iloc[:,i], Depth)
            figure.set_title(f"{well_name[j]}")
            figure.set_xlabel(log.columns[i])  
            ax[j][0].set_ylabel('Depth')
            figure.set_ylim(min(Depth), max(Depth))
            figure.invert_yaxis()
            figure.grid()
            
logs_plot_per_well(EBDdf_2)
```

## Cross-plot all well logs for all wells

```python
well_name = EBDdf_2["WELL"].unique().tolist()
property_list = list(EBDdf_2.columns)
property_list.remove('WELL')

def cross_plot(well, property1, property2, property3):

    '''
    Function to cross-plot well logs
    '''
    plt.scatter(EBDdf_2[str(property1)][EBDdf_2['WELL']==str(well)],EBDdf_2[str(property2)][EBDdf_2['WELL']==str(well)],
                c=EBDdf_2[property3][EBDdf_2['WELL']==str(well)],cmap=plt.cm.jet)
    plt.title(f"{property1} vs. {property2} color_coded with {property3} at well {well}")
    plt.xlabel(f"{property1}")
    plt.ylabel(f"{property2}")
    plt.legend([str(property3)],loc='upper right')
    plt.colorbar()
    plt.grid()
    plt.show()
```

```python
@interact(property=widgets.SelectionSlider(options=property_list, description='Colour_Code'))
def interactive_colorcoded_crossplot(property, Well_Name=well_name, X_Axis=EBDdf_2, Y_Axis=EBDdf_2):   
    cross_plot(Well_Name, X_Axis, Y_Axis, property)
```
